#### Perform spatial join on puma shapefile with uber ride lat/lon

Final Output I want is as follows:


```python
__RIDE_FROM_PUMA__ = sjoin(uber_aug14_gdf, __RAW_PUMA_SHP_GDF__, how='left',op="within")
__RIDE_FROM_PUMA__.head()
```

```python
def process_dt(s):
    dt, t = tuple(s.split())
    mo, d, y = tuple([int(e) for e in dt.split('/')]) 
    h, m, s = tuple([int(e) for e in t.split(':')])
    return datetime(year=y, month=mo, day=d, hour=h, minute=m, second=s)
__RIDE_FROM_PUMA__['dt'] = __RIDE_FROM_PUMA__['Date/Time'].apply(process_dt)
__RIDE_FROM_PUMA__.head()
```

##### I need these columns for groupby time dim

```python
__RIDE_FROM_PUMA__['date'] = __RIDE_FROM_PUMA__['dt'].apply(lambda ts: ts.date())
__RIDE_FROM_PUMA__['hour'] = __RIDE_FROM_PUMA__['dt'].apply(lambda ts: ts.hour)
```

##### groupby time dimension with granularity of an hour, and count rides

```python
__RIDE_FROM_PUMA_GRPBY__ = __RIDE_FROM_PUMA__.groupby(['puma', 'date', 'hour']).count()
```

```python
__RIDE_FROM_PUMA_GRPBY__.drop(labels=[c for c in __RIDE_FROM_PUMA_GRPBY__.columns[1:]], axis=1, inplace=True)
__RIDE_FROM_PUMA_GRPBY__.head()
```

##### Iterate through all PUMAs and Merge the respective puma's hourly rideship count to the DataFrame RIDE_CNT_BY_PUMA

```python
RIDE_CNT_BY_PUMA = pd.DataFrame(
    index=[datetime(year=2014, month=8, day=1, hour=0) + timedelta(days=h//24, hours=h%24) for h in range(24 * 31)]
)
for puma in __RIDE_FROM_PUMA_GRPBY__.index.levels[0]:
    _df = __RIDE_FROM_PUMA_GRPBY__.loc[puma]
    _ix = [ts + relativedelta(hour=hr) for ts, hr in _df.index]
    _df.index = _ix
    _df.rename(mapper={'Date/Time': int(puma)}, inplace=True, axis=1)
    RIDE_CNT_BY_PUMA = RIDE_CNT_BY_PUMA.merge(_df, how='left', left_index=True, right_index=True)

RIDE_CNT_BY_PUMA = RIDE_CNT_BY_PUMA.fillna(0)
for c in RIDE_CNT_BY_PUMA.columns:
    RIDE_CNT_BY_PUMA[c] = RIDE_CNT_BY_PUMA[c].apply(int)
RIDE_CNT_BY_PUMA.head()
```

```python
ax = RIDE_CNT_BY_PUMA.sum(axis=0).plot(
    kind='bar'
    , color='red'
    , title ="Number of Rides in Aug. 2014"
    , figsize=(15,10)
    , fontsize=12
)
```
